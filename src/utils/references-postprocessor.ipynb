```python
html_content = """
<div class="csl-entry" id="ref-c1-baertschi-mulligan:2001" role="listitem"> 
    <span class="smallcaps">Baertschi</span>, Bernard and 
    <span class="smallcaps">Mulligan</span>, Kevin, eds. 2001a. 
    <em><span>Les nationalismes</span></em>. 
    <span><span>É</span>thique et philosophie morale</span>. 
    Paris: Presses Universitaires de France. 
</div>
"""

# Parse the HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Find all spans with the class "smallcaps"
smallcaps_spans = soup.find_all("span", class_="smallcaps")

# Find the last smallcaps span
if smallcaps_spans:
    last_smallcaps_span = smallcaps_spans[-1]
    # Get the next span after the last smallcaps
    next_span = last_smallcaps_span.find_next("span")
    # Extract the title text
    if next_span:
        title = next_span.get_text(strip=True)
        print("Title:", title)
    else:
        print("No span found after the last smallcaps.")
else:
    print("No smallcaps spans found.")
```

```python

html_content = '<div class="csl-entry" id="ref-c1-baertschi-mulligan:2001" role="listitem"> <span class="smallcaps">Baertschi</span>, Bernard and <span class="smallcaps">Mulligan</span>, Kevin, eds. 2001a. <em><span>Les nationalismes</span></em>. <span><span>É</span>thique et philosophie morale</span>. Paris: Presses Universitaires de France. </div>'

soup = BeautifulSoup(html_content, "html.parser")

all_ems = soup.find_all("em")

for em in all_ems:
    em_tag = em.find("span")
    if em_tag:
        title = em_tag.get_text(strip=True)
        print("Title:", title)
        break
else:
    print("No <span> tag found explicitly within a <em>.")
```
